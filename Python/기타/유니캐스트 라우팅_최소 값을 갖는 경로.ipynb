```python
# 유니캐스트 라우팅 알고리즘 구현
# 최소 값을 가지는 경로를 찾아 출력하는 기능을 수행
# 목적지 고정일때

#주어진 행렬은 다음과 같음 
#    A B C D E F G
# A  0 2 0 3 0 0 0
# B  2 0 5 0 4 0 0
# C  0 5 0 0 0 4 3
# D  3 0 0 0 5 0 0
# E  0 4 0 5 0 2 0
# F  0 0 4 0 2 0 1
# G  0 0 3 0 0 1 0

import copy as cp # 깊은 복사 연산(deepcopy()) 사용해야 할까? 안 쓰고 copy()는 안되지 당연히


# 그래프를 딕셔너리형으로 변환
# 딕셔너리의 각 알파벳(key)은 딕셔너리형 값(value)을 가짐
# 값을 표현하는 딕셔너리는 연결된 노드와 그 값에 대응되는 cost값을 뜻함
# 즉 연결된 노드 이름을 key로, 그 경로에 대한 cost를 value로 가짐

graph = {
        'A' : {'B':2, 'D':3},
        'B' : {'A':2, 'C':5, 'E':4},
        'C' : {'B':5, 'F':4, 'G':3},
        'D' : {'A':3, 'E':5},
        'E' : {'B':4, 'D':5, 'F':2},
        'F' : {'C':4, 'E':2, 'G':1},
        'G' : {'C':3, 'F':1}
        }

dest = 'G' # 목적지는 G로 고정
#dest = input("목적지 입력(A~G)")

path = {} # 빈 딕셔너리형 변수 path 선언 - 최소거리값, 경로, 방문여부 추가

for i in graph.keys() :# graph의 key를 이용한 반복문 수행
    # 최소거리값(cost), 경로(리스트형 : 지나간 노드를 표시해야 함), 방문여부
    # key로 받아서 출력하기 위함
    path[i] = {'최소 Cost 값' : 0, '경로' : [], '방문여부' : 0}
    
path
```

```python
# 함수 정의 : 매개변수가 하나인 visit_place 함수
# 방문여부를 1로 카운트하여 방문했음을 체크
# 연결여부를 확인하여 연결이 되어 있으면
def visit_place(visit) : 
    path[visit]['방문여부'] = 1
    if visit == dest :
        break
    # graph에서 매개변수로 들어온 값에 대한 key/value 쌍(items)으로 반복
    # visit
    for node, cost in graph[visit].items() : 
        sum_cost = path[visit]['최소 Cost 값'] + cost # 경로값 누적
        
        # 조건 두가지 : 최소 cost값이 기존 cost값보다 크거나 경로에 없다면
        if (path[node]['최소 Cost 값'] >= sum_cost) or not path[node]['경로'] :
            path[node]['최소 Cost 값'] = sum_cost # 기존 cost값을 바꾸고
            #path[node]['경로'] = path[visit]['경로'].copy() # copy()와 deepcopy() 차이?
            # copy()의 경우 얕은 복사 : 원본을 변경할 경우 같이 변경
            # deepcopy()의 경우 깊은 복사 연산 수행 : 객체를 새로 생성, 별개의 존재로 보며 원본 변경해도 변경되지 않음
            path[node]['경로'] = cp.deepcopy(path[visit]['경로'])
            path[node]['경로'].append(visit)
                       
#path
```

```python
#source = input("출발지 입력(A~G) : ")    # 최초 출발지 노드 값 입력받음
#visit_place(source) # 출발지로 입력받은 값을 매개변수로 하여 함수 실행

# cost값이랑 경로 알맞게 잘 들어갔나?
#path 
```

```python
source = input("출발지 입력(A~G)")    # 최초 출발지 노드 값 입력받음

visit_place(source) # 출발지로 입력받은 

while 1 : # loop - 연결되는 노드 없을 때까지 반복(1)
    # path.values()와 key에서 가장 큰 항목
    next_cost = max(path.values(), key = lambda x:x['최소 Cost 값'])['최소 Cost 값']
    next_visit = ''    # 채우지 않음 

    for next_node, search in path.items():
        if 0 < search['최소 Cost 값'] <= next_cost and not search['방문여부']:
            next_cost = search['최소 Cost 값']
            next_visit = next_node

    if next_visit == '': # 다음 연결되는 노드가 없을 경우
        print (next_visit) # 출력 후
        break # 반복문 빠져나가기
    
    visit_place(next_visit) # 그 다음 노드와 연결되어 있는지 없는지 확인하여 cost 또는 빈칸으로 채워넣음
    # 다음 연결된 노드를 매개변수로 하여 함수 실행(재귀)

print ("(", source, "부터", dest,"까지의 경로)")
print ("방문 노드 경로 : ", path[dest]['경로'])
print ("목적지까지의 경로 중 최소값(Cost) : ", path[dest]['최소 Cost 값'])
```
