# 1-1 : 시계열 데이터_여러 데이터

```python
# 주식 이름과 종목 코드
# LG전자(066570), 삼성전자(005930), LG디스플레이(034220)
# import문과 start/end 지정문은 생략하였음

##### LG전자, 삼성전자, LG디스플레이의 주식 시계열 데이터를 가져와 데이터프레임으로 저장 #####
LGE = data.get_data_yahoo('066570.KS', start, end) # 1. LG전자의 주식 시계열 데이터를 가져와 데이터프레임으로 저장
SAM = data.get_data_yahoo('005930.KS', start, end) # 2. 삼성전자의 주식 시계열 데이터를 가져와 데이터프레임으로 저장
LGD = data.get_data_yahoo('034220.KS', start, end) # 2. LG디스플레이의 주식 시계열 데이터를 가져와 데이터프레임으로 저장
##### LG전자, 삼성전자, LG디스플레이의 주식 시계열 데이터를 가져와 데이터프레임으로 저장 : 끝 #####
```

```python
##### 가져온 데이터를 그래프로 그리기 #####
### 데이터프레임(LG)에는 open, high, low 등 여러 열 값이 존재
### 하나의 값만 추려서 그래프로 그리려고 함

LGE['Close'].plot(figsize = (15, 10), grid = True); # LGE의 close 열만 가지고 그래프 만들기
SAM['Close'].plot(figsize = (15, 10), grid = True); # SAM의 close 열만 가지고 그래프 만들기
LGD['Close'].plot(figsize = (15, 10), grid = True); # LGD의 close 열만 가지고 그래프 만들기

# 그래프의 가로/세로 크기 지정-15,10, 격자 보이기
# 하나라도 grid 옵션을 ()안에 기입하지 않을 경우 격자 활성화되지 않음
# figsize는 하나의 ()안에만 기입해도 적용됨

plt.title("각 종목의 일자별 주가 추이") # 그래프의 제목(타이틀) 지정 : LG전자
plt.xlabel("일자") # 그래프의 x축 이름 지정 : 일자
plt.ylabel("주가") # 그래프의 y축 이름 지정 : 주가
plt.legend(loc = "best")

plt.show() # 그래프 보이게 하기
##### 가져온 데이터를 그래프로 그리기 : 끝#####
```

```python
##### 가져온 데이터를 그래프로 그리기_2 #####
### 데이터프레임(LG)에는 open, high, low 등 여러 열 값이 존재
### 하나의 값만 추려서 그래프로 그리려고 함

### 위 코드와의 차이점 : 위의 코드는 범례 활성화시 모두 close로만 보이므로 어떤 종목인지 구분 불가능
### 각 열의 close 이름을 종목:close로 변경하여 구분할 수 있게 하기 위함


##### 각 종목 데이터의 원본을 복사하여 새로운 데이터프레임을 생성 #####
### Close에 해당하는 열의 이름만 각 종목의 이름으로 바꿔줌
### 기존 종목 데이터는 뒤에서도 이용하기 때문에, 불편함을 줄이기 위해 원본을 건드리지 않고 복사본을 활용하는 방식을 이용하였음
### 위의 코드 대비 추가된 부분임 : 어떤 종목이 어떤 색인지 구분하기 위해

LGE_1 = LGE.copy() #원본 복사
LGE_1['LG전자'] = LGE['Close']

SAM_1 = SAM.copy() #원본 복사
SAM_1['삼성전자'] = SAM['Close']

LGD_1 = LGD.copy() #원본 복사
LGD_1['LG디스플레이'] = LGD['Close']

##### 각 종목 데이터의 원본을 복사하여 새로운 데이터프레임을 생성 : 끝 #####


LGE_1['LG전자'].plot(figsize = (15, 10), grid = True); # LGE의 close 열만 가지고 그래프 만들기
SAM_1['삼성전자'].plot(figsize = (15, 10), grid = True); # SAM의 close 열만 가지고 그래프 만들기
LGD_1['LG디스플레이'].plot(figsize = (15, 10), grid = True); # LGD의 close 열만 가지고 그래프 만들기

# 그래프의 가로/세로 크기 지정-15,10, 격자 보이기
# 하나라도 grid 옵션을 ()안에 기입하지 않을 경우 격자 활성화되지 않음
# figsize는 하나의 ()안에만 기입해도 적용됨

plt.title("각 종목의 일자별 주가 추이") # 그래프의 제목(타이틀) 지정 : LG전자
plt.xlabel("일자") # 그래프의 x축 이름 지정 : 일자
plt.ylabel("주가") # 그래프의 y축 이름 지정 : 주가
plt.legend(loc = "best")

plt.show() # 그래프 보이게 하기
##### 가져온 데이터를 그래프로 그리기_2 : 끝#####
```
